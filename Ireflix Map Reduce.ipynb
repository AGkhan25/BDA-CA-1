```python
#create a data set for viewing 
#import liability 
import pandas as pd
from collections import defaultdict
```

```python
#import data in notebook 
data=pd.read_csv("Ireflix_viewing_dataset.csv")
```

```python
# view data
data.head()
```

```python
# Write a mapper() function that reads each line and outputs (genre, 1) key-value pairs
# to write a mapper we do need to define a function

def map_function(data):
    mapped = []
    for genre in data['Genre']:
        mapped.append((genre, 1))
    return mapped

# Write a reducer() function that groups by genre and returns the total view count per genre.
#To write a reduce function we will first sort the data and then write reduce function

def shuffle_sort(mapped_data):
    grouped_data = defaultdict(list)

    for key, value in mapped_data:
        grouped_data[key].append(value)

    return grouped_data


def reduce_function(grouped_data):
    reduced_data = {}

    for key, values in grouped_data.items():
        reduced_data[key] = sum(values)

    return reduced_data

# Write a main() function that calls both functions and prints the final output.

def get_top_8_items(reduced_data):
    return sorted(
        reduced_data.items(),
        key=lambda item: item[1],
        reverse=True
    )[:8]  

    
```

```python
# Run your solution
#run map function
mapped=map_function(data)
```

```python
#print the map 
mapped
```

```python
# run groupinp and sort
grouped= shuffle_sort(mapped)
```

```python
#print group
grouped 
```

```python
# run reduce function
reduced=reduce_function(grouped)
```

```python
# print reduce
reduced
```

```python
# get top 8 results 
Top_8= get_top_8_items(reduced)
```
